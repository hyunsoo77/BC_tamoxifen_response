# drug effect

```r
idx <- which(list_gmt[["unc_lccc"]]$term == "T47D_control_vs_tamoxifen_down_Tumor cells")
genes <- list_gmt[["unc_lccc"]][idx, "gene"]

genes <- genes[genes %in% rownames(mtx)]

rna$sum_t47d_down_gene_expr <- colSums(mtx[genes,])
rna$mean_t47d_down_gene_expr <- colMeans(mtx[genes,])

f_tumor_epi <- grepl(pattern_tumor_epi, rna$cell.type)
log2fc_t47d_down_genes <- rep(NA, ncol(rna))
for (sample in names(list_samples)) {
    
    sample_control <- list_samples[[sample]]$control
    sample_treatment <- list_samples[[sample]]$treatment
    
    idx_control <- which(rna$Sample == sample_control & f_tumor_epi)
    idx_treatment <- which(rna$Sample == sample_treatment & f_tumor_epi)
    
    mean_control <- mean(rna$mean_t47d_down_gene_expr[idx_control], na.rm = TRUE)
    mean_treatment <- mean(rna$mean_t47d_down_gene_expr[idx_treatment], na.rm = TRUE)
    
    cat(sprintf("%s: treatment: %g, control: %g\n", sample_control, mean_treatment, mean_control))
    log2fc_t47d_down_genes[idx_treatment] <- log2(rna$mean_t47d_down_gene_expr[idx_treatment] / mean_control)
    
} # for

rna$log2fc_t47d_down_genes <- log2fc_t47d_down_genes


```
